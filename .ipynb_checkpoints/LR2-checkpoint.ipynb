```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, LabelEncoder
from sklearn.preprocessing import MinMaxScaler, StandardScaler

# Загрузка датасета Titanic
titanic = sns.load_dataset('titanic')

print("Датасет Titanic (первые 5 строк):")
print(titanic.head())

print("\nИнформация о датасете:")
print(titanic.info())
```

```python
# Вывод количества пропусков
print("Количество пропусков в каждом столбце:")
print(titanic.isnull().sum())

# Визуализация пропусков
plt.figure(figsize=(10, 5))
sns.heatmap(titanic.isnull(), cbar=False, yticklabels=False, cmap='viridis')
plt.title('Карта пропусков в датасете Titanic')
plt.show()
```

```python
# ### Метод 1: Удаление строк с пропусками (для колонки embarked)
# В колонке 'embarked' всего 2 пропуска. Их можно удалить без потери данных.

# %%
titanic_cleaned = titanic.copy()
print(f"Размер до удаления: {titanic_cleaned.shape}")
titanic_cleaned = titanic_cleaned.dropna(subset=['embarked'])
print(f"Размер после удаления пропусков в 'embarked': {titanic_cleaned.shape}")
print(f"Пропуски в 'embarked' после удаления: {titanic_cleaned['embarked'].isnull().sum()}")
```

```python
# ### Метод 2: Заполнение пропусков (для колонки age)
# Возраст пропущен у многих пассажиров. Заполним медианным значением.

# %%
# Способ 1: Pandas fillna
titanic_filled = titanic_cleaned.copy()
median_age = titanic_filled['age'].median()
titanic_filled['age'] = titanic_filled['age'].fillna(median_age)
print(f"Возраст заполнен медианой: {median_age}")
print(f"Пропуски в 'age' после fillna: {titanic_filled['age'].isnull().sum()}")
```

```python
# Способ 2: SimpleImputer из sklearn (для демонстрации)
# Вернем пропуски обратно для примера
titanic_with_nan = titanic_cleaned.copy()
imputer = SimpleImputer(strategy='median')
titanic_with_nan[['age']] = imputer.fit_transform(titanic_with_nan[['age']])
print(f"Пропуски в 'age' после SimpleImputer: {titanic_with_nan['age'].isnull().sum()}")
```

```python
# ## 2. Кодирование категориальных признаков

# Подготовим чистый датасет для кодирования
df_encoded = titanic_filled.copy()
```

```python
# ### Label Encoding (для бинарных признаков)
# Пол пассажира: male/female преобразуем в 0 и 1.

# %%
le_sex = LabelEncoder()
df_encoded['sex_encoded'] = le_sex.fit_transform(df_encoded['sex'])
print("Label Encoding для 'sex':")
print(df_encoded[['sex', 'sex_encoded']].drop_duplicates())
```

```python
# ### One-Hot Encoding (для номинативных признаков без порядка)
# Порт посадки (embarked): значения 'S', 'C', 'Q'. Между ними нет иерархии.

# %%
# Метод pandas get_dummies
embarked_dummies = pd.get_dummies(df_encoded['embarked'], prefix='embarked')
df_final = pd.concat([df_encoded, embarked_dummies], axis=1)
print("One-Hot Encoding для 'embarked' (первые 5 строк):")
print(df_final[['embarked', 'embarked_C', 'embarked_Q', 'embarked_S']].head())
```

```python
# Альтернатива с использованием sklearn OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)
embarked_encoded = encoder.fit_transform(df_encoded[['embarked']])
embarked_df = pd.DataFrame(embarked_encoded, columns=encoder.get_feature_names_out(['embarked']))
print("\nOneHotEncoder из sklearn (первые 5 строк):")
print(embarked_df.head())
```

```python
# ### Ordinal Encoding (для порядковых признаков)
# Класс пассажира (class): 'First', 'Second', 'Third'. Имеет явный порядок.

# %%
# Задаем правильный порядок классов
class_order = [['First', 'Second', 'Third']]  # Список списков для OrdinalEncoder
ord_enc = OrdinalEncoder(categories=class_order)
df_final['class_encoded'] = ord_enc.fit_transform(df_final[['class']])
print("Ordinal Encoding для 'class' с пользовательским порядком:")
print(df_final[['class', 'class_encoded']].drop_duplicates().sort_values('class_encoded'))
```

```python
# ## 3. Масштабирование данных

# Выделим числовые признаки для масштабирования
# Возьмем возраст (age) и стоимость билета (fare)
numeric_features = df_final[['age', 'fare']].copy()
print("Исходные числовые данные (первые 5 строк):")
print(numeric_features.head())

print("\nСтатистика ДО масштабирования:")
print(numeric_features.describe())
```

```python
# ### StandardScaler (Стандартизация)
# Преобразует данные так, чтобы среднее = 0, стандартное отклонение = 1.

# %%
scaler_std = StandardScaler()
scaled_std = scaler_std.fit_transform(numeric_features)
scaled_std_df = pd.DataFrame(scaled_std, columns=['age_std', 'fare_std'])

print("Данные после StandardScaler (первые 5 строк):")
print(scaled_std_df.head())
print("\nСтатистика ПОСЛЕ StandardScaler (mean ~0, std ~1):")
print(scaled_std_df.describe().round(2))
```

```python
# ### MinMaxScaler (Нормализация)
# Приводит данные к заданному диапазону [0, 1].

# %%
scaler_mm = MinMaxScaler()
scaled_mm = scaler_mm.fit_transform(numeric_features)
scaled_mm_df = pd.DataFrame(scaled_mm, columns=['age_mm', 'fare_mm'])

print("Данные после MinMaxScaler (первые 5 строк):")
print(scaled_mm_df.head())
print("\nСтатистика ПОСЛЕ MinMaxScaler (min=0, max=1):")
print(scaled_mm_df.describe().round(2))
```

```python
# Визуализация сравнения масштабирования
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

axes[0].scatter(numeric_features['age'], numeric_features['fare'], alpha=0.5)
axes[0].set_title('Оригинальные данные')
axes[0].set_xlabel('Age')
axes[0].set_ylabel('Fare')

axes[1].scatter(scaled_std_df['age_std'], scaled_std_df['fare_std'], alpha=0.5)
axes[1].set_title('StandardScaler')
axes[1].set_xlabel('Age (стандартиз.)')
axes[1].set_ylabel('Fare (стандартиз.)')

axes[2].scatter(scaled_mm_df['age_mm'], scaled_mm_df['fare_mm'], alpha=0.5)
axes[2].set_title('MinMaxScaler')
axes[2].set_xlabel('Age (нормализ.)')
axes[2].set_ylabel('Fare (нормализ.)')

plt.tight_layout()
plt.show()
```

```python
# ### Итоговый датасет
# Объединим все преобразованные признаки в один DataFrame

# %%
final_dataset = pd.concat([
    df_final[['survived', 'pclass', 'sex', 'age', 'sibsp', 'parch', 'fare', 'class']],
    scaled_std_df,
    scaled_mm_df,
    df_final[['sex_encoded', 'class_encoded']],
    embarked_dummies
], axis=1)

print("Итоговый датасет после всей обработки (первые 5 строк):")
print(final_dataset.head())
print(f"\nРазмер итогового датасета: {final_dataset.shape}")
print(f"Количество пропусков в итоговом датасете: {final_dataset.isnull().sum().sum()}")
```
